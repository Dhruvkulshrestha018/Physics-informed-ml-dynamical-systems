```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example setup (you should replace these with your real data if available)
Nx, Nt = 256, 100
x = np.linspace(-1, 1, Nx)
t = np.linspace(0, 1, Nt)

# Dummy solution for demo (replace this with your real Burgers' solution)
# Here, we'll assume a manufactured solution like:
u = np.outer(np.sin(np.pi * x), np.exp(-t))  # shape (Nx, Nt)

# Create meshgrid
X, T = np.meshgrid(x, t)  # X and T shape will be (Nt, Nx), same as u.T

# Plotting the surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, T, u.T, cmap='viridis')  # Transpose u to match shape
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u(x,t)')
ax.set_title("3D Surface Plot of Burgers' Equation Dataset")
plt.tight_layout()
plt.show()
```
